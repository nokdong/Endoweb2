```python
from tqdm import tqdm
pt=0
endo=0
patient_nums=[]
exams=Exam.objects.all()
for exam in tqdm(exams):
    if exam.hospital_no in patient_nums:
        
        already_pt = Patient.objects.get(hospital_no = exam.hospital_no)
        endoscopy=Endoscopy(patient=already_pt, date=exam.exam_date, type=exam.exam_type, doc=exam.exam_doc, 
                            source=exam.exam_class, place=exam.exam_place, 
                            Dx=exam.exam_Dx, procedure=exam.exam_procedure, Bx_result=exam.Bx_result,
                            followup_period=exam.follow_up, re_visit_call=exam.phone_check, re_visit=exam.re_visit, 
                            Bx_result_call=exam.Bx_result_call, re_visit_date="1900-01-01")
        if exam.hospital_no in temp_dict:
            setattr(endoscopy, 'Bx_result_call', temp_dict[exam.hospital_no])
            #print ("Bx_result_call2", endoscopy.Bx_result_call)
        #print ("patient:", already_pt.name)
        #print ("endo:", endoscopy.date)
        #already_pt.save()
        endoscopy.save()
    else :   
        patient = Patient(name = exam.patient_name, hospital_no=exam.hospital_no, sex=exam.patient_sex, birth=exam.patient_birth, phone=exam.patient_phone, address=".")
        patient.save()
        patient_nums.append(patient.hospital_no)
        endoscopy=Endoscopy(patient=patient, date=exam.exam_date, type=exam.exam_type, doc=exam.exam_doc, source=exam.exam_class, place=exam.exam_place, 
                            Dx=exam.exam_Dx, procedure=exam.exam_procedure, Bx_result=exam.Bx_result, Bx_result_call=exam.Bx_result_call,
                            followup_period=exam.follow_up, re_visit_call=exam.phone_check, re_visit=exam.re_visit, re_visit_date="1900-01-01")
        if exam.hospital_no in temp_dict:
            setattr(endoscopy, 'Bx_result_call', temp_dict[exam.hospital_no])
            #print ("Bx_result_call2", endoscopy.Bx_result_call)
        #print ("patient:", patient.name, "endo:", endoscopy.date)
        #patient.save()
        endoscopy.save()
    
```

```python
def add_month(date, months):
    month = date.month + int(months) - 1
    year = int(date.year + (month / 12))
    month = (month % 12) + 1
    day = date.day
    new_date = date.replace(year=year, month=month, day=1)
    return new_date
```

```python
for endoscopy in endoscopy_list:
    endoscopy.followup_date = add_month(endoscopy.date, endoscopy.followup_period)
    endoscopy.save()
```
